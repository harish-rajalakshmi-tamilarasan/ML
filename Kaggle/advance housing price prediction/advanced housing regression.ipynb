```python
#function to see the missing data in a dataframe
def missing_data(df,n):    
    total = df.isnull().sum().sort_values(ascending=False)          # Total No of missing values
    percentage = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)*100  # % of Missing values
    No_unique_val = df.nunique()# No of unique values
    dtype = df.dtypes
    missing_data = pd.concat([total, percentage, No_unique_val,dtype], axis=1, 
                             keys=['Total No of missing val', '% of Missing val','No of unique val','dtype'], sort = False)
    
    print(missing_data.head(n))
```

```python
missing_data(train_df,20)
missing_data(test_df,30)
```

```python
def find_corelation_two_columns(df,a,b):
    filtered_df = df[df[a].isnull()]
    value_counts = filtered_df[b].value_counts()

    # Plot the bar graph
    value_counts.plot(kind='bar')
    plt.xlabel('Values in column A')
    plt.ylabel('Count')
    plt.title('Bar Graph of column A where column B is empty')
    plt.show()
    
def find_corelation_two_columns_hist(df,a,b):
    filtered_df = df[df[b].isnull()]
    filtered_df[a].plot(kind='hist', bins=10, edgecolor='black')
    plt.xlabel('Values in column A')
    plt.ylabel('Frequency')
    plt.title('Histogram of column A where column B is empty')
    plt.show()
```
